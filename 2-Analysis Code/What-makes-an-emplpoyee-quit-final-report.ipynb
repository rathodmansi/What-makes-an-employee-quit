### Research Question 3: Setting up Logistic Regression

Research Question: Can we predict employee attrition beforehand. Our hypothesis behind this research is that predicting employee attrition beforehand will help us take corrective measures and retain high performing employees.

In the following lines of code, I have first performed dummy encoding to convert the categorical variables to numerical daat types and then divided the dataset into train and test group.

```python
## One-hot encoding
emp_data2 = pd.get_dummies(emp_data)
emp_data2.head()
emp_data2.shape

## Dividing data into train and test dataset
X = emp_data2.drop(['left'], axis=1)
y = emp_data2['left']
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, random_state=123)

## Setting up Logistic Regression Model
print("Setting up Logistic Regression Model")
log_model = LogisticRegression()
fit = log_model.fit(X_train, y_train)

## Looking into the coefficients values for our different features
print("The columns are", X.columns, "\nThe coefficients are", fit.coef_)
print("")
coeff = fit.coef_
```

Conclusions from RQ 3
 1. Satisfaction level is highly significant in causing an employee to quit
 2. Work accident is also inversely affecting attrition (not able to understand why)
 3. Promotion in last 5 yrs is negatively affecting attrition, ie if not promoted then likely to leave
 4. HIGH salary is inversely related to the attrition
 5. Use the probabilities of employee leaving the comapany to tackle the employees
 
 
<b> Now we look into the accuracy of our logistic regression model and see how it is performing on test dataset</b>

```python
## Looking into the accuracy of model 
log_model.score(X_test, y_test)
print("Thus, the accuracy of our model is 79%, i.e. out of 10 people quitting, it is able to identify approximately 8 people")
```

```python
## Looking into the probabilities of employee leaving
attrition_prob = log_model.predict_proba(X_test)
attrition_prob
```

#### The above vector now contains the probabilities of employees quitting their jobs. This information can be used to segregate the emplpyees into different categories such as high-risk attrition and low-risk attrition. Following up, then we can take corresponding actions to help increase these emplpyees satisfaction level and thus retain them within the organization.
